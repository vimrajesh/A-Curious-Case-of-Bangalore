```python
new= merged[["date","aod","evaptrans","fire","State"]]
len(new.State.unique())
```

```python
new.corr()
```

```python
# fig, axs = plt.subplots(nrows=3, figsize = (20,10))
plt.figure(figsize = (20,10))
alol = new.groupby("date").mean()
plt.plot(alol.index.tolist(),alol.aod/max(alol.aod), label = "aod")
plt.plot(alol.index.tolist(),alol.evaptrans/max(alol.evaptrans),label = "evaptrans")
plt.plot(alol.index.tolist(),alol.fire/max(alol.fire),label = "fire")
plt.legend()
# axs[0].plot(alol.index.tolist(),alol.aod, label = "aod")
# axs[0].set_title("AOD")
# axs[1].plot(alol.index.tolist(),alol.evaptrans,label = "evaptrans")
# axs[1].set_title("EVAPTRANS")
# axs[2].plot(alol.index.tolist(),alol.fire,label = "fire")
# axs[2].set_title("Fire")
```

```python
# alol = alol.resample('Y').agg('mean')
alol1 = alol.resample('Y').agg('mean')
alol1["year"] = alol1.index.year
fig, axs = plt.subplots(nrows=3, figsize = (20,10))
axs[0].plot(alol1.year,alol1.aod, label = "aod")
axs[0].set_title("AOD")
axs[1].plot(alol1.year,alol1.evaptrans,label = "evaptrans")
axs[1].set_title("EVAPTRANS")
axs[2].plot(alol1.year,alol1.fire,label = "fire")
axs[2].set_title("Fire")
```
